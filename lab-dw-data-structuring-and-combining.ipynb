```python
import pandas as pd
file1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'

df1 = pd.read_csv(file1)


df1.columns = df1.columns.str.lower()
df1.columns = df1.columns.str.replace(' ', '_')
df1.rename(columns = {'st':'state'}, inplace = True)


df1.head()
```

```python
df1.dtypes
#Need to change dtype of: gender -> category, customer_lifetime_value -> float64, number_of_open_complaints -> float64, 
#policy_type -> category, vehicule_class -> category.
gender = {
    'Femal': 'F',
    'Male': 'M', 
    'female': 'F'
}
df1['gender'] = df1['gender'].replace(gender)
df1['gender'] = df1['gender'].astype('category')
df1.dtypes
df1.head()
```

```python
df1['customer_lifetime_value'] = df1['customer_lifetime_value'].str.replace('%','')
df1['customer_lifetime_value'].unique()
df1['customer_lifetime_value'] = df1['customer_lifetime_value'].astype('float64')
df1['customer_lifetime_value'].dtype
```

```python
df1['number_of_open_complaints'].unique()
def split(row):
    if isinstance(row, str):
        return row.split('/')[1]
    else:
        return 0

df1['number_of_open_complaints'] = df1['number_of_open_complaints'].apply(split)
df1['number_of_open_complaints'].unique()
df1['number_of_open_complaints'] = df1['number_of_open_complaints'].astype(int)

```

```python
df1['policy_type'].unique()
df1['policy_type'] = df1['policy_type'].astype('category')
df1['policy_type'].dtype
```

```python
df1['vehicle_class'].unique()
df1['vehicle_class'] = df1['vehicle_class'].astype('category')
df1['vehicle_class'].dtype
```

```python
df1['education'].replace('Bachelors', 'Bachelor')
```

```python
file2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df2 = pd.read_csv(file2)
df2.columns = df2.columns.str.lower()
df2.columns = df2.columns.str.replace(' ', '_')
df2.head()
```
